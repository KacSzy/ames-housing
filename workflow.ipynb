- Diagnoza brakow

```
# 2.2 Procent brakow w kolumnach
length = df.shape[0]

for col in df.columns:
	null_percent = df[col].isna().sum() / length * 100
	if null_percent > 0:
		print(f'{col}: {null_percent:.2f}%')
```

```
# 2.2 Uzupelnienie brakow w kolumnach kategorycznych
for col in cat_cols:
	df[col] = df[col].fillna('None')
```

```
# 2.2 Diagnostyka brakow w wierszach
row_null_counts = df.isna().sum(axis=1)

for n in range(1, 11):
	amount = (row_null_counts >= n).sum()
	if amount > 0:
		print(f"Rows with > {n} missing values: {amount}")
```

```
# 2.2 Procent wierszy z >= 2 brakami
print(f'Percentage of rows with missing values: {(row_null_counts >= 2).sum()/length * 100:.2f}%')
```

```
# 2.2 Kontrola brakow po uzupelnieniach
for col in df.columns:
	null_percent = df[col].isna().sum() / length * 100
	if null_percent > 0:
		print(f'{col}: {null_percent:.2f}%')
```

```
# 2.2 Uzupelnienie Garage Yr Blt na podstawie Year Built
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(df['Year Built'])
```

```
# 2.2 Uzupelnienie Lot Frontage srednia w Neighborhood
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda x: x.fillna(x.mean()))
```

```
# 2.2 Kontrola brakow po uzupelnieniu Lot Frontage
for col in df.columns:
	null_percent = df[col].isna().sum() / length * 100
	if null_percent > 0:
		print(f'{col}: {null_percent:.2f}%')
```

```
# 2.2 Uzupelnienie Lot Frontage srednia globalna
df['Lot Frontage'] = df['Lot Frontage'].fillna(df['Lot Frontage'].mean())
```

```
# 2.2 Uzupelnienie Mas Vnr Area wartoscia 0
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)
```

```
# no value => no basement, no garage... => 0
cols_to_zero = [
    'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF',
    'Bsmt Full Bath', 'Bsmt Half Bath', 'Garage Cars', 'Garage Area'
]

df[cols_to_zero] = df[cols_to_zero].fillna(0)
```

```
# 2.2 Finalna kontrola brakow
for col in df.columns:
	null_percent = df[col].isna().sum() / length * 100
	if null_percent > 0:
		print(f'{col}: {null_percent:.2f}%')
```

```
# 2.4 Zakodowanie kaegorycznych
'''
Target encoding, also known as mean encoding, involves replacing each category in a categorical variable with the mean (or other aggregated statistic) of the target variable for that category. It leverages the relationship between the categorical feature and the target variable, providing a way to encode categorical information directly into numeric form
~ https://sagarikakathuria29.medium.com/understanding-the-difference-between-target-encoding-and-frequency-encoding-1d9bd264b8e
'''

cat_cols = df.select_dtypes(exclude=np.number).columns

global_mean = df['SalePrice'].mean()

for col in cat_cols:
    # Map each category to mean SalePrice
    mean_map = df.groupby(col)['SalePrice'].mean()
    df[f"{col}_te"] = df[col].map(mean_map).fillna(global_mean)

df.drop(columns=cat_cols, inplace=True)

print(df.select_dtypes(exclude="number").shape[1])
print(df.select_dtypes(include="number").shape[1])
```

### 2.5 Redukcja wymiarowości

Z uwagi na rozmiar danych, w projekcie zostaną zastosowane trzy metody redukcji wymiarowości:
1. Incremental PCA
2. Random Projection
3. K-Means

Wynik działania każdego z algorytmów (dane w pełni gotowe do modelowania) zostanie zapisany pod odpowiednią ścieżką:\
*data_prepared/Ames_{algorithm}.csv*

W kolejnych etapach porównam wyniki osiągnięte przez poszczególne modele na każdym zestawie danych. Kluczowym elementem będzie dobór odpowiednich hiperparametrów dla każdego z algorytmów, aby wyłonić najbardziej optymalne rozwiązanie.

#### 2.5.1 Incremental PCA


```
from pathlib import Path
from sklearn.decomposition import IncrementalPCA
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['SalePrice'])
y = df['SalePrice']

X_scaled = StandardScaler().fit_transform(X)

ipca = IncrementalPCA(n_components=X.shape[1], batch_size=256)
X_pca = ipca.fit_transform(X_scaled)

pca_columns = [f"PC{i + 1}" for i in range(X_pca.shape[1])]
pca_df = pd.DataFrame(X_pca, columns=pca_columns, index=df.index)
pca_df['SalePrice'] = y

output_dir = Path('data_prepared')
output_dir.mkdir(parents=True, exist_ok=True)
pca_df.to_csv(r'data_prepared/Ames_pca.csv', index=False)
```

#### 2.5.2 Random Projection

#### 2.5.3 K-Means


```
from sklearn.cluster import KMeans

X = df.drop(columns=['SalePrice'])
y = df['SalePrice']

X_scaled = StandardScaler().fit_transform(X)

inertia_values = []

K_range = range(2, 21) # 2-20

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_) # zwraca sumę odlegnosci od cendroidów

plt.figure(figsize=(15, 5))
plt.plot(K_range, inertia_values, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

```
from sklearn.metrics import silhouette_score

scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)

    score = silhouette_score(X_scaled, cluster_labels)
    scores.append(score)

plt.figure(figsize=(15, 5))
plt.plot(K_range, scores, marker='s', color='green', linestyle='-')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

Po przeprowadzeniu analizy z wykorzystaniem Elbow Method oraz Silhouette Score, podjąłem decyzję o odrzuceniu algorytmu K-Means jako metody redukcji wymiarowości dla tego zbioru. Wykres inercji nie wykazał jednoznacznego punktu przegięcia, a globalne maksimum Silhouette Score wyniosło zaledwie 0.14. Tak niskie wartości metryk dowodzą, że zbiór Ames Housing ma charakter "ciągły". Narzucanie danym sztucznego podziału klastrowego spowodowałoby stratę informacji.

```
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


def optimize_models(X_train, X_test, y_train, y_test, model, param_grid, dataset_name):
    max_cols = X_train.shape[1]
    k_to_test = [10, 15, 20, 25, 30, 35, 40, 45, 50, max_cols]

    best_mae = float('inf')
    best_results = {}
    best_k = None
    best_params = None

    for k in k_to_test:
        X_train_k = X_train.iloc[:, :k]
        X_test_k = X_test.iloc[:, :k]

        grid = GridSearchCV(
            model,
            param_grid,
            cv=5,
            scoring='neg_mean_squared_error',
            n_jobs=-1
        )

        grid.fit(X_train_k, y_train)

        best_current_model = grid.best_estimator_
        y_pred = best_current_model.predict(X_test_k)

        mae = mean_absolute_error(y_test, y_pred)

        if mae < best_mae:
            best_mae = mae
            best_k = k
            best_params = grid.best_params_
            best_results = {
                'Dataset': dataset_name,
                'Model': model.__class__.__name__,
                'Optimal_K': k,
                'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
                'MAE': mae,
                'R2': r2_score(y_test, y_pred)
            }

    print(f"\n--- {dataset_name} + {model.__class__.__name__} ( {best_mae:.2f} )---")
    print(f"  k: {best_k}")
    for key, value in best_params.items():
        print(f"  {key}: {value}")

    return best_results
```

```
def optimize_models(X_train, X_test, y_train, y_test, model, param_grid, dataset_name, n):
    grid = GridSearchCV(
        model,
        param_grid,
        cv=5,
        scoring='neg_mean_squared_error',
        n_jobs=-1
    )

    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    best_params = grid.best_params_

    results = {
        'Dataset': dataset_name,
        'Model': model.__class__.__name__,
        'Optimal_K': n,
        'RMSE': rmse,
        'MAE': mae,
        'R2': r2
    }

    print(f"\n--- {dataset_name} + {model.__class__.__name__} + MAE: {mae:.2f} ---")
    for key, value in best_params.items():
        print(f"  {key}: {value}")

    return results
```

```
def plot_top_3_models(df):
    fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
    fig.suptitle('Najlepsze modele według błędu MAE', fontsize=18, y=1.05)

    df_unique = df.sort_values('MAE').drop_duplicates(subset=['Model', 'Dataset_Group'])

    # 1. Top 3 PCA
    top_pca = df_unique[df_unique['Dataset_Group'] == 'PCA'].nsmallest(3, 'MAE')
    sns.barplot(data=top_pca, x='Model', y='MAE', hue='Model', ax=axes[0], palette='Blues_d', legend=False)
    axes[0].set_title('Top 3: PCA', fontsize=14)
    axes[0].set_ylabel('Mean Absolute Error (MAE)')

    # 2. Top 3 Random Projection
    top_rp = df_unique[df_unique['Dataset_Group'] == 'Random Projection'].nsmallest(3, 'MAE')
    sns.barplot(data=top_rp, x='Model', y='MAE', hue='Model', ax=axes[1], palette='Oranges_d', legend=False)
    axes[1].set_title('Top 3: Random Projection', fontsize=14)
    axes[1].set_ylabel('')

    # 3. Top 3 Ogólnie
    top_overall = df_unique.nsmallest(3, 'MAE').copy()
    top_overall['Model_Dataset'] = top_overall['Model'] + " (" + top_overall['Dataset_Group'].str.replace('Random Projection', 'RP') + ")"
    sns.barplot(data=top_overall, x='Model_Dataset', y='MAE', hue='Model_Dataset', ax=axes[2], palette='Greens_d', legend=False)
    axes[2].set_title('Top 3: Klasyfikacja Generalna', fontsize=14)
    axes[2].set_ylabel('')

    for ax in axes:
        ax.set_xlabel('')
        for container in ax.containers:
            ax.bar_label(container, fmt='%.0f', padding=3, fontsize=11)

    plt.tight_layout()
    plt.show()


def plot_performance_heatmap(df):
    plt.figure(figsize=(10, 6))

    pivot_df = df.pivot_table(index='Model', columns='Dataset_Group', values='MAE', aggfunc='min')

    sns.heatmap(pivot_df, annot=True, fmt=".0f", cmap="viridis_r", linewidths=.5, cbar_kws={'label': 'MAE'})

    plt.title('MAE Heatmap', fontsize=15, pad=15)
    plt.ylabel('')
    plt.xlabel('Algorytm Redukcji Wymiarowości')
    plt.tight_layout()
    plt.show()


plot_df = df.copy()

plot_df['Dataset_Group'] = plot_df['Dataset'].apply(
    lambda x: 'Random Projection' if str(x).startswith('Random Projection') else x
)

plot_top_3_models(plot_df)
plot_performance_heatmap(plot_df)
```
